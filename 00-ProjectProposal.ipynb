## Research Question

How does borrower income, employment length, home ownership status, and intent of loan affect the probability of a loan being defaulted?

The study focuses on the correlation between the binary outcome variable loan_status, where loan_status =1 indicates a borrower has failed to pay off (defaulted), while the loan_status= 0 indicates repayment as agreed. Utilizing numerous descriptive statistics, such as with mean and median, the study attempts to summarize borrower and loan characteristics from the data set as well as use group comparisons via calculating the default rate across the categories of home ownership and loan intent to better be able to identify difference in default risk between groups while applying logistic regression, to model to the audience the probability of loan default as a function to person_income, person_emp_length,person_home_ownership, and loan_intent and whether the variables play a factor in the probability of the binary outcome variable loan_status being defaulted.

## Background and Prior Work


Credit risk assessment is the fundamental process used by financial institutions to predict the likelihood that a borrower will fail to meet their debt obligation. Interestingly, in the modern day economy characterized by fluctuating interest rates and shifting employment trends, lenders have moved away from simple credit score checks toward comprehensive risk modeling. This process involves analyzing a combination of “capacity” (a borrower’s ability to pay) and “stability” (their profession and their residential history). As financial service becomes increasingly digitized in 2026, the industry is focusing on identifying which specific borrower attributes such as annual income and employment tenure act as the most reliable anchors for financial responsibility [1](https://blog.crsoftware.com/credit-risk-decisioning-in-2025-from-manual-reviews-to-smart-automation).

Previous work in this field has shown that housing status and the specific intent of a loan are major predictors of default behavior (the failure of a borrower to fulfill their legal obligations to repay a debt according to agreed-upon terms). Research into consumer default has consistently demonstrated that homeowners are statistically less likely to default compared to renters, as property ownership suggests a level of financial stability and collateral that mitigates risk [2](https://pmc.ncbi.nlm.nih.gov/articles/PMC9041570/). Furthermore previous data science projects utilizing credit risk datasets have discovered that "loan intent" acts as a crucial qualitative filter; for instance, loans intended for education or home improvement often have different repayment trajectories than those intended for high-interest debt consolidation. What these earlier projects really showed is that you can't just look at a person’s background by itself; you have to look at their demographics alongside the actual reason they’re taking out the loan in the first place.

Data from the Consumer Financial Protection Bureau (CFPB) shows that a borrower’s age is a major factor in how they handle debt, especially as they transition through different life stages. Their research highlights that younger borrowers often have much higher delinquency rates, likely because they have shorter credit histories and less financial experience compared to older generations/older borrowers [3](https://www.consumerfinance.gov/data-research/consumer-credit-trends/student-loans/lending-borrower-age/). By looking at these age-based trends, we can see that risk isn't just about how much money someone makes, but also where they are in their life and their age. This is really important for our project because it helps explain why younger people who might still be in the early stages of their careers are often flagged as higher risk even if they have a steady income.

To model these outcomes, researchers predominantly utilize Logistic Regression. While more complex machine learning algorithms exist, recent studies published in early 2025 emphasize that Logistic Regression is the preferred tool for credit risk because it provides a "probability score" that is easily interpretable for regulatory audits [4](https://doi.org/10.9734/ajpas/2025/v27i2710). This method allows analysts to see exactly how much an additional year of employment or a $10,000 increase in income reduces the probability of a binary "default" outcome. Our project builds on this existing research by using these established statistical methods to determine the predictive power of income, employment, and housing status.

## Hypothesis


Borrowers with lower income, higher loan-to-income ratios, shorter employment length, past defaults and shorter credit history are more likely to default on loans.

Borrowers' financial stability and credit scores influence their likelihood of default on loans. Lower income, higher loan-to-income ratios, shorter employment length, prior defaults and shorter credit history indicate higher financial risk. All those factors together indicate that borrowers are more likely to miss loan payments.
