## Research Question

How does borrower income, employment length, home ownership status, and intent of loan affect the probability of a loan being defaulted?

The study focuses on the correlation between the binary outcome variable loan_status, where loan_status =1 indicates a borrower has failed to pay off (defaulted), while the loan_status= 0 indicates repayment as agreed. Utilizing numerous descriptive statistics, such as with mean and median, the study attempts to summarize borrower and loan characteristics from the data set as well as use group comparisons via calculating the default rate across the categories of home ownership and loan intent to better be able to identify difference in default risk between groups while applying logistic regression, to model to the audience the probability of loan default as a function to person_income, person_emp_length,person_home_ownership, and loan_intent and whether the variables play a factor in the probability of the binary outcome variable loan_status being defaulted.

## Hypothesis


Borrowers with lower income, higher loan-to-income ratios, shorter employment length, past defaults and shorter credit history are more likely to default on loans.

Borrowers' financial stability and credit scores influence their likelihood of default on loans. Lower income, higher loan-to-income ratios, shorter employment length, prior defaults and shorter credit history indicate higher financial risk. All those factors together indicate that borrowers are more likely to miss loan payments.


## Data

The ideal dataset we would want to answer the research question would have to contain a significantly large amount of data entries (at least 10,000) since a loan default is relatively uncommon. Within each group of loan statuses, we need quantitative and categorical data labels like income, expenses, house ownership status, loan intents, loan amounts, interest rates, age, etc associated with every entry in the dataset. The data would be collected at the time of loan origination and through self reported entries that are verified by tax documents to ensure the validity of the data. The data would consists of entries from people from ages 20 to 40 who have income. This data would be stored in a structured format like SQL database or CSV file with every row corresponding to a person's data entry and every column representing data labels.

**Credit Risk Dataset**
   This dataset is located on Kaggle (https://www.kaggle.com/datasets/laotse/credit-risk-dataset). It is an open source dataset that is free to use for our purpose and does not require any permission requirements. The important variables in this dataset that we might use are ```person_income```, ```loan_percent_income```, ```loan_int_rate```, ```person_home_ownership```. These variables will help us create predictive models by utilizing such features related to repayment ability and context for a person's risk assesment. This dataset has around 32,000 entries and is well documented with details regarding rows and columns.

**Credit Scoring Dataset Creditworthiness Prediction**
   This dataset is located on Kaggle (https://www.kaggle.com/datasets/junaid512/credit-scoring-dataset-creditworthiness-prediction).
